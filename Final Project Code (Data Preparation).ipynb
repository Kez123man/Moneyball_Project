```python
# FBref link for Web Scraping and Data Extraction

# Firstly, a variable is created using the FBref website link 
url_df_FB = 'https://fbref.com/en/comps/Big5/2023-2024/stats/players/2023-2024-Big-5-European-Leagues-Stats'

# Next, the "pd.read_html()" function is used to extract the data from FBref's HTML table
# Table is extracted at "Index[0]"
df_FB = pd.read_html(url_df_FB)[0]
df_FB.head()

# Adapted From:
# Title: How to scrape and personalize data from FBref with Python: A guide to unlocking Football Insights
# Author: Ricardo Andre
# Date: 2023
# Code Version: N/A
# Source: "https://medium.com/@ricardoandreom/how-to-scrape-and-personalize-data-from-fbref-with-python-a-guide-to-unlocking-football-insights-7e623607afca"


# Title: 2023-2024 Big 5 European Leagues: Standard Stats.
# Author: FBref
# Date: 2024
# Code Version: N/A
# Source: "https://fbref.com/en/comps/Big5/2023-2024/stats/players/2023-2024-Big-5-European-Leagues-Stats"
```

```python
# Now, this code creates a list with the new column names
# Removes unwanted parts of the column name such as "0_level_0"
new_col_FB = []
for new_FB in df_FB.columns:
    if 'level_0' in new_FB:
        new_col_name_FB = new_FB.split()[-1]
        new_col_FB.append(new_col_name_FB)  
    else:
        new_col_FB.append(new_FB)  

# This part renames the columns
df_FB.columns = new_col_FB
df_FB = df_FB.fillna(0)

df_FB.head()

# Adapted From:
# Title: How to scrape and personalize data from FBref with Python: A guide to unlocking Football Insights
# Author: Ricardo Andre
# Date: 2023
# Code Version: N/A
# Source: "https://medium.com/@ricardoandreom/how-to-scrape-and-personalize-data-from-fbref-with-python-a-guide-to-unlocking-football-insights-7e623607afca"

# Adapted From:
# Author: Open AI
# Date: 2024
# Source: ChatGPT [online] "https://chat.openai.com/"
```
